```python
# 提取vs ph gr 共有的点
import numpy as np
import matplotlib.pyplot as plt
import os
from scipy.interpolate import interp1d
filepath_ph = "./dispersion_ph/"
filepath_gr = "./dispersion_gr/"
filepath_vs = './InputVsData/'
#layer thickness
depth_spacing=0.5# 1,2,5,6,10
depth_max = 150
if depth_spacing in [0.5,1,2,5,6,10]:
    depth = np.arange(0,depth_max+depth_spacing,depth_spacing)
os.system('rm -rf Vs_GroundTruth && mkdir Vs_GroundTruth')
os.system('rm -rf layers_vs && mkdir layers_vs')
print("depth: ",depth)
```

```python
depth_visual = np.array([0,5,10,15,20,30,40,60,80,100,120,depth_max])
lay1 = [];lay2 = [];lay3 = [];lay4 = [];lay5 = [];lay6 = [];lay7 = [];
lay8 = [];lay9 = [];lay10 = [];lay11 = [];lay12 = [];lay13 = [];
count = 0
for lat in np.arange(0,180,0.1):
    for lon in np.arange(0,360,0.1):
        key = "%.2f"%lat+ "_"+"%.2f"%lon
        key_name = key + '.txt'
        file_vs = filepath_vs + key_name
        if  os.path.exists(file_vs):
            temp = np.loadtxt(file_vs); 
            if len(temp)>1:
                count =count +1
                print(count,key_name)
                depth_temp=temp[:,0]; vs_temp = temp[:,1]
                fl = interp1d(depth_temp, vs_temp, kind='slinear')
                vs = fl(depth)
                # for using those data to predict dispersion curve comparing the vs from cnn
                vs_out = np.array([depth,vs])
                np.savetxt("./Vs_GroundTruth/"+key_name,vs_out.T,fmt="%10.5f")
                # end
 
                vs = fl(depth_visual)
                lay1.append([lon,lat,vs[0]])
                lay2.append([lon,lat,vs[1]])
                lay3.append([lon,lat,vs[2]])
                lay4.append([lon,lat,vs[3]])
                lay5.append([lon,lat,vs[4]])
                lay6.append([lon,lat,vs[5]])
                lay7.append([lon,lat,vs[6]])
                lay8.append([lon,lat,vs[7]])
                lay9.append([lon,lat,vs[8]])
                lay10.append([lon,lat,vs[9]])
                lay11.append([lon,lat,vs[10]])
                lay12.append([lon,lat,vs[11]])
```

```python
all_lyers =[lay1,lay2,lay3,lay4,lay5,lay6,lay7,lay8,lay9,lay10,lay11,lay12];
fig=plt.figure(1,figsize=(6,4))
count=0
cm=[[2.0,3.6],[2.8,3.6],[3.1,3.7],[3.2,3.8],[3.2,3.9],[3.2,4.2],[3.3,4.7],[3.6,4.7],[4.1,4.7],[4.2,4.8],[4.1,4.8],[4.1,4.8],[4.1,4.7]]
from mpl_toolkits.basemap import Basemap
#     0         5         10         15       20        30          40       60         80         100      120      150       180
for lay in all_lyers: 
    count =count+1
    m = Basemap(projection='mill', llcrnrlon=73, llcrnrlat=18, urcrnrlon=135,
            urcrnrlat=53)
    m.drawcountries(linewidth=1.5)
    m.drawcoastlines(linewidth=1.5)
    x,y = m(lay[:,0],lay[:,1])
    z = lay[:,2]
    sc=plt.scatter(x,y,c=z,s=15,cmap='jet_r')       
    plt.title('Depth= '+str(depth_visual[count-1])+'km')
    cb=plt.colorbar(sc,shrink=0.75)
    cb.ax.set_title('Vs(km/s)')
    # draw parallels
    m.drawparallels(np.arange(10,90,10),labels=[1,0,0,1])
    # draw meridians
    m.drawmeridians(np.arange(-180,180,10),labels=[1,1,0,1])
#     plt.clim(cm[count-1])
    plt.pause(0.1)
    plt.savefig( './Figs_vs/sws-'+str(depth_visual[count-1])+'.jpg',bbox_inches='tight') 
    fig.clear()
```
